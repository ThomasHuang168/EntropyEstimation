# 3 Uniform RV in (0,1)

```python
low, high, RVsize, numRV = 0, 1, 1000, 6
depend = np.array([0, 1])
rv_c = DC.getRandomVar_select(np.random.uniform, low, high, RVsize, numRV, depend)
```

```python
from sklearn.metrics import mean_squared_error
def logMSEscorer(clf, X, y):
    y_est = clf.predict(X)
    return mean_squared_error(y, y_est)
```

```python
def varEntropy(y):
    return np.var(y)
```

```python
from sklearn.neighbors import KNeighborsRegressor
regressor = KNeighborsRegressor(n_neighbors=numRV)

CVFold = 3
Entropy = DC.computeEnt(rv_c, regressor,logMSEscorer, varEntropy, CVFold, True)
```
